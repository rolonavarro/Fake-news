```python
# Import basic libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import gzip
```

```python
test_df = pd.DataFrame(pd.read_csv('../00_Resources/fake_or_real_news.csv'))
test_df = test_df.drop(columns=['Unnamed: 0'])
test_df = test_df.replace({'FAKE': 'False', 'REAL': 'True'})
test_df
```

```python
with open('../00_Resources/true_fake.csv', 'rb') as fd:
    gzip_fd = gzip.GzipFile(fileobj=fd)
    df = pd.read_csv(gzip_fd)
df = df.drop(columns={'Unnamed: 0','subject','date'})
df = df.rename(columns={'category':'label'})
# df = df.append(test_df, ignore_index=True)
df = df.sample(frac=1).reset_index(drop=True)
# df = df.loc[df['label'] != 'unknown']
df
```

```python
df_url = pd.DataFrame(pd.read_csv('../00_Resources/fake_news_with_url.csv'))
df_url = df_url.rename(columns={'Headline':'title','Body':'text'})
df_url['label'] = df_url['Label'].replace({0: 'False', 1: 'True'})
df_url = df_url.drop(columns=['URLs','Label'])
df_url = df_url.dropna(how='any')
df_url
```

```python
df1 = pd.DataFrame(pd.read_csv('../00_Resources/Fake-News/train.csv'))
df2 = pd.DataFrame(pd.read_csv('../00_Resources/Fake-News/test.csv'))
df3 = pd.DataFrame(pd.read_csv('../00_Resources/Fake-News/submit.csv'))

df4 = df1.append(df2, ignore_index=True)
df_merge = pd.merge(df3, df4, on='id')

df_merge['label'] = df_merge['label_x'].replace({1: 'False', 0: 'True'})
df_merge = df_merge.drop(columns=['id','label_x','author','label_y'])
df_merge = df_merge.dropna(how='any')

df_merge
```

```python
df2 = df.append(test_df, ignore_index=True)
df3 = df2.append(df_url, ignore_index=True)
df4 = df3.append(df_merge, ignore_index=True)

df5 = df4.sample(frac=1).reset_index(drop=True)

df5
```

```python
df5.to_csv('../00_Resources/all_merged_datasets.csv', compression='gzip')
```
